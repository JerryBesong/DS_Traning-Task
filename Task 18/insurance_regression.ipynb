In this task, i am asked to create a scatter plot using "insurance.csv" as my dataset.

lets first import all of the packages for this task.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
# import and read ("insurance.csv") file
data = pd.read_csv("insurance.csv")
data.head()
```

```python
# using iloc to assign our variables to our x and y axis
x = data.iloc[:, 0].values
y = data.iloc[:, 6].values

x = x.reshape(-1,1)

plt.scatter(x, y, color = "blue")
plt.title("Age VS Charges Relationship.")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
# fitting of a model using linear regression
model = LinearRegression()
model.fit(x, y)
```

```python
# predict on x using another age value of 70
# Make predictions on the same x values
y_predict = model.predict(x)

plt.scatter(x, y, color = "blue")
plt.plot(x, y_predict, color = "r")
plt.title("Age VS Charges Relationship.")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()
```

```python
predic = [[70]]

x_predict = np.append(x, predic).reshape(-1, 1)
y_predict = model.predict(x_predict)

# plot graph
plt.scatter(x, y, color = "blue")
plt.plot(x_predict, y_predict, color = "r")
plt.title("Age VS Charges Relationship.")
plt.xlabel("Age")
plt.ylabel("Charges")
plt.show()

print(f"for a prediction of age {predic}, the charges will be {model.predict(predic)}")
```
