```python
# Import pandas
import pandas as pd
```

```python
# Create a DataFrame with balance.txt
df = pd.read_csv("balance.txt", sep = " " )  #delim_whitespace=True)
```

```python
df.head()
```

* Compare the average income based on ethnicity.

```python
# first lets know how many unique in ethnicity 
group = df.loc[:,"Ethnicity"].unique()
print(group)
```

```python

print("The average income based on ethnicity is", df[df.Ethnicity =="Caucasian"].loc[:,"Income"].mean())
print("The average income based on ethnicity is", df[df.Ethnicity =="Asian"].loc[:,"Income"].mean())
print("The average income based on ethnicity is", df[df.Ethnicity =="African American"].loc[:,"Income"].mean())
```

* On average, do married or single people have a higher balance?

```python
print("The average income of married based in our dataframe is", df[df.Married =="Yes"].loc[:,"Balance"].mean())
print("The average income of married based in our dataframe is", df[df.Married =="No"].loc[:,"Balance"].mean())

print("Our output shows that unmarried people have a little bit of higher balance.")
```

* What is the highest income in our dataset?

```python
print("The highest income in our dataset is", df.loc[:, "Income"].max())
```

* What is the lowest income in our dataset?

```python
print("The highest income in our dataset is", df.loc[:, "Income"].min())
```

* How many cards do we have recorded in our dataset? (Hint: use `sum()`)

```python
# using .iloc function
print("The numbers of card in my data base is", df.iloc[:,4].sum())
```

* How many females do we have information for vs how many males? 

```python
print("The Numbers of females we have their informations is", df["Gender"].value_counts()["Female"])
print("The Numbers of Males we have their informations is", df["Gender"].value_counts()["Male"])

```
